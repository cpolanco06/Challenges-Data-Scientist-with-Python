### Contexto
#### Entre 1901 y 2016, los Premios Nobel y el Premio de Ciencias Económicas fueron entregados 579 veces a 911 personas y organizaciones. El Premio Nobel es un premio internacional administrado por la Fundación Nobel en Estocolmo, Suecia, y basado en la fortuna de Alfred Nobel, inventor y empresario sueco. En 1968, Sveriges Riksbank estableció el Premio Sveriges Riksbank en Ciencias Económicas en memoria de Alfred Nobel, fundador del Premio Nobel. Cada premio consta de una medalla, un diploma personal y un premio en efectivo.
#### Una persona u organización galardonada con el Premio Nobel se denomina Premio Nobel. La palabra "laureado" se refiere al significado de la corona de laurel. En la antigua Grecia, se otorgaban coronas de laurel a los vencedores como señal de honor.
### Contenido
#### Este conjunto de datos incluye un registro de cada individuo u organización que recibió el Premio Nobel desde 1901 hasta 2016.
### Reto: ¿Qué país ha ganado más premios en cada categoría?

```python
#Importamos todas las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Funciones para automatizar pasos

```python
#Funcion para grafico de recuento
def count_plot(x, data, height, width, xlabel, ylabel, title, h_title, order=None):
    g = sns.catplot(x=x, data=data,
                    kind="count",
                    order=order,
                    height=height, aspect=width)
    g.set(xlabel=xlabel,
          ylabel=ylabel)
    g.fig.suptitle(title, y=h_title)
    return plt.show()
```

```python
#Funcion para grafico de barras
def bar_plot(x, y, data, height, width, xlabel, ylabel, title, h_title, orient='v'):
    g = sns.catplot(x=x, y=y, data=data,
                    kind="bar",
                    orient=orient,
                    height=height, aspect=width)
    g.set(xlabel=xlabel,
          ylabel=ylabel)
    g.fig.suptitle(title, y=h_title)
    return plt.show()
```

```python
#Funcion para hacer el recuento por paises ganadores en determinada columna de categoria y toma los primeros 10 valores
def func_count(column_name, input_name_df, category,  output_name_df):
    global chemistry_countries_top10, economics_countries_top10, literature_countries_top10, medicine_countries_top10
    global peace_countries_top10, physics_countries_top10
    DF = input_name_df.groupby(column_name).agg({column_name: 'count'})
    DF = DF.rename(columns={column_name: 'Nobel Prizes Won_' + category})
    DF = DF.sort_values(by='Nobel Prizes Won_' + category, ascending=False)
    DF_top10 = DF[:10]
    output_name_df = DF_top10
    output_name_df = output_name_df.sort_values(by='Nobel Prizes Won_' + category)
    output_name_df = output_name_df.reset_index()
    return output_name_df
```

#### Solución a la pregunta del reto: ¿Qué país ha ganado más premios en cada categoría?

```python
#Dataframes para cada categoria
chemistry_df = nobel[(nobel.Category == 'Chemistry')]
economics_df = nobel[(nobel.Category == 'Economics')]
literature_df = nobel[(nobel.Category == 'Literature')]
medicine_df = nobel[(nobel.Category == 'Medicine')]
peace_df = nobel[(nobel.Category == 'Peace')]
physics_df = nobel[(nobel.Category == 'Physics')]
```

```python
#Dataframes vacios para almacenar los dataframes resultantes del recuento para cada categoria
chemistry_countries_top10 = pd.DataFrame()
economics_countries_top10 = pd.DataFrame()
literature_countries_top10 = pd.DataFrame()
medicine_countries_top10 = pd.DataFrame()
peace_countries_top10 = pd.DataFrame()
physics_countries_top10 = pd.DataFrame()
```

```python
#Recuento de los paises ganadores por cada categoria
chemistry_countries_top10 = func_count('Birth Country', chemistry_df, 'Chemistry', chemistry_countries_top10)
economics_countries_top10 = func_count('Birth Country', economics_df, 'Economics', economics_countries_top10)
literature_countries_top10 = func_count('Birth Country', literature_df, 'Literature', literature_countries_top10)
medicine_countries_top10 = func_count('Birth Country', medicine_df, 'Medicine', medicine_countries_top10)
peace_countries_top10 = func_count('Birth Country', peace_df, 'Peace', peace_countries_top10)
physics_countries_top10 = func_count('Birth Country', physics_df, 'Physics', physics_countries_top10)
```

```python
#Grafico de barras de los paises que han ganado más premios en cada categoría
fig, ax = plt.subplots(2,3, figsize=(20,15))
plt.subplots_adjust(wspace = 0.6)

ax[0,0].barh(y='Birth Country', width='Nobel Prizes Won_Chemistry', data=chemistry_countries_top10, label='Chemistry',
             color=sns.color_palette('Purples', len(range(10))))
ax[0,1].barh(y='Birth Country', width='Nobel Prizes Won_Economics', data=economics_countries_top10, label='Economics',
             color=sns.color_palette('Reds', len(range(10))))
ax[0,2].barh(y='Birth Country', width='Nobel Prizes Won_Literature', data=literature_countries_top10, label='Literature',
             color=sns.color_palette('Oranges', len(range(10))))
ax[1,0].barh(y='Birth Country', width='Nobel Prizes Won_Medicine', data=medicine_countries_top10 , label='Medicine',
             color=sns.color_palette('Blues', len(range(10))))
ax[1,1].barh(y='Birth Country', width='Nobel Prizes Won_Peace', data=peace_countries_top10, label='Peace',
             color=sns.color_palette('Greys', len(range(10))))
ax[1,2].barh(y='Birth Country', width='Nobel Prizes Won_Physics', data=physics_countries_top10, label='Physics',
             color=sns.color_palette('Greens', len(range(10))))

ax[0,0].set_xlabel('Count')
ax[0,1].set_xlabel('Count')
ax[0,2].set_xlabel('Count')
ax[1,0].set_xlabel('Count')
ax[1,1].set_xlabel('Count')
ax[1,2].set_xlabel('Count')

ax[0,0].set_ylabel('Country')
ax[0,1].set_ylabel('Country')
ax[0,2].set_ylabel('Country')
ax[1,0].set_ylabel('Country')
ax[1,1].set_ylabel('Country')
ax[1,2].set_ylabel('Country')

plt.suptitle('Countries with the most nobel prizes won by category (Top 10)', fontsize=20)

ax[0,0].legend()
ax[0,1].legend()
ax[0,2].legend()
ax[1,0].legend()
ax[1,1].legend()
ax[1,2].legend()
```

### Observamos que a partir de la grafica y de acuerdo al recuento de la columna "Birth Country" por cada categoria que el pais con mas premios ganados por categoria es "United States of America" solo supererado en la categoria de Literatura por "France".
#### Otros paises que estan el top 3 de cada categoria son "Germany", "United Kingdom", "Norway" y "Sweden".
